```python
#1:Days to Seconds Converter
# The input() function returns a string, so we convert it to an integer using int().
days = int(input("Enter the number of days: "))

# Constants
HOURS_IN_DAY = 24
MINUTES_IN_HOUR = 60
SECONDS_IN_MINUTE = 60

# Number of seconds.
seconds = days * HOURS_IN_DAY * MINUTES_IN_HOUR * SECONDS_IN_MINUTE

# Print the final result in a user-friendly format.
print(f"{days} days is equal to {seconds} seconds. ‚è±Ô∏è")
```

```python
#2: Sphere Volume Calculator

# Import the 'math' module to get access to the value of Pi.
import math

# We use float() to allow for decimal values in the radius.
radius = float(input("Enter the radius of the sphere: "))

# Formula is V = (4/3) * œÄ * r^3.
# We use the exponential operator (**) to calculate the cube of the radius as requested.
volume = (4/3) * math.pi * (radius ** 3)

# Print the calculated volume.
print(f"The volume of the sphere with radius {radius} is {volume:.2f}. üåê")
```

```python
#3: Square Area and Perimeter Calculator using Function

def calculate_square_properties(side_length):
    """
    This function calculates the area and perimeter of a square.
    Args:
        The length of one side of the square.
    Returns:
        The calculated area and perimeter.
    """
    #Area: side * side
    area = side_length ** 2
    #Perimeter: 4 * side
    perimeter = 4 * side_length
    # Return both calculated values
    return area, perimeter

# Convert the input to a floating-point number.
side = float(input("Enter the side length of the square: "))

# Call the function with the user's input and unpack the returned tuple into two variables.
square_area, square_perimeter = calculate_square_properties(side)

# Display the results to the user.
print(f"\nFor a square with side length {side}:")
print(f"The area is: {square_area} üü©")
print(f"The perimeter is: {square_perimeter}")
```

```python
#4: Character Case Checker using Functions

def check_character_case(char_input):
    """
    This function checks if a given character is uppercase, lowercase, or neither.
    It prints the result directly.
    Arguments:
        char_input (str): The character to be checked.
    """
    # Use the isupper() string method to check if the character is uppercase.
    if char_input.isupper():
        print(f"The character '{char_input}' is uppercase.")
    # If not uppercase, use the islower() method to check if it's lowercase.
    elif char_input.islower():
        print(f"The character '{char_input}' is lowercase.")
    # If it is neither, it is not an alphabet character.
    else:
        print(f"The input '{char_input}' is not an alphabet character (e.g., a number or symbol).")

# We take the first character of the input string using [0] to ensure we check only one.
user_char = input("Enter a single character: ")[0]

# Call the function to perform the check and print the result.
check_character_case(user_char)
```

```python
#5. Pseudocode to Python Conversion

# SET x TO 0, y TO 20
x = 0
y = 20

# The pseudocode's REPEAT-UNTIL loop runs as long as the condition (y < 6) is false.
# This is equivalent to a 'while' loop that runs as long as the opposite condition (y >= 6) is true.
while y >= 6:
    # SUBTRACT 4 FROM y
    y -= 4
    # ADD 2/y TO x
    # Note: This operation happens after 'y' has been updated.
    x += (2 / y)

# DISPLAY x
print(f"The final value of x is: {x}")
print(f"The loop terminated because the final value of y is {y}, which is less than 6.")
```

```python
#6.Calculate the average of 5 user-input values

# Initialize an empty list to store the numbers.
values = []

print("Please enter 5 numbers to calculate their average.")

# Use a for loop that repeats 5 times to get the user's input.
# range(5) generates numbers from 0 to 4.
for i in range(5):
    # Prompt the user for a number.
    # We use float() to convert the input string to a number that can have decimals.
    number = float(input(f"Enter number {i + 1}: "))

    # Add the entered number to our list.
    values.append(number)

# Calculate the average after the loop is finished.
# The sum() function adds up all elements in the list.
# The len() function gets the count of elements in the list.
average = sum(values) / len(values)

# Display the list of numbers entered and their calculated average.
print(f"\nThe numbers you entered are: {values}")
print(f"The average of these numbers is: {average} üìä")
```
